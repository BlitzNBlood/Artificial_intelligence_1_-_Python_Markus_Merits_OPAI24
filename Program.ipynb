```python

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
seqIdList = list()
seqList = list()
seqData = list()
with open("dna_raw.txt","r") as file:
    count = 0
    for line in file: #takes each line seperatley
        if (line.__contains__(">")): #detects and removes > and puts the line which had the > sign into seqIdList for later use
            remove = ">"
            for char in remove:
                line = line.replace(char, "")
            seqIdList.append(line)
        else:
            low = line.lower() #makes all letters lower case
            print(low)
            a = low.count("a")
            t = low.count("t")
            c = low.count("c")
            g = low.count("g")
            if (count == 0): #first run needs to make index 0 - 3
                seqData.append(a)
                seqData.append(t)
                seqData.append(c)
                seqData.append(g)
            else: #replaces values so barplot doesnt use old values and new ones at the same time
                seqData[0] = a
                seqData[1] = t
                seqData[2] = c
                seqData[3] = g
            count = count + 1
            df = pd.DataFrame({'a': [seqData[0]], 't': [seqData[1]], 'c': [seqData[2]], 'g': [seqData[3]]}) #sets processed data into dataframe (dictionary)
            ax = sns.barplot(data=df) #makes a barplot that uses df as a dataframe
            plt.xlabel('DNA Letters') #adds "DNA Letters to x label"
            plt.ylabel('Proficiency') #adds "Proficiency to y label"
            plt.title(seqIdList[count - 1]) #adds correct sequence name to correct sequence data
            plt.show()
            #seqList.append(line) #puts sequence data into list for further processing

    '''count = 0
    for row in seqList:
        low = row.lower()
        a = low.count("a")
        t = low.count("t")
        c = low.count("c")
        g = low.count("g")
        if (count == 0):
            seqData.append(a)
            seqData.append(t)
            seqData.append(c)
            seqData.append(g)
        else:
            seqData[0] = a
            seqData[1] = t
            seqData[2] = c
            seqData[3] = g
        count = count + 1
        df = pd.DataFrame({'a': [seqData[0]], 't': [seqData[1]], 'c': [seqData[2]], 'g': [seqData[3]]}) #sets processed data into dataframe
        ax = sns.barplot(data=df) #makes a barplot that uses df as a dataframe
        plt.xlabel('DNA Letters') #adds "DNA Letters to x label"
        plt.ylabel('Proficiency') #adds "Proficiency to y label"
        plt.title(seqIdList[count - 1]) #adds correct sequence name to correct sequence data
        plt.show()'''
    
        
```
